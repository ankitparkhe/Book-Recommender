```python
@interact
def recomd_books_publisher(publisher_name=list(df['publisher'].value_counts().index)):
    a = df[df['publisher'] == publisher_name][['title', 'average_rating']]
    a = a.sort_values(by='average_rating', ascending=False)
    return a.head(10)
```

```python
# Sort upon Authors 

@interact
def recomd_books_authors(authors_name=list(df['authors'].value_counts().index)):
    a = df[df['authors'] == authors_name][['title', 'average_rating']]
    a = a.sort_values(by='average_rating', ascending=False)
    return a.head(10)
```

```python
df.columns
```

```python
# Sorting Upon Language 

@interact
def recomd_language_code(language_code=list(df['language_code'].value_counts().index)):
    a = df[df['language_code'] == language_code][['title', 'average_rating']]
    a = a.sort_values(by='average_rating', ascending=False)
    return a.head(10)
```

### Data Preprocessing 

```python
def num_to_obj(x):
    if 0 < x < 1:
        return "between 0 and 1"
    elif 1 < x <= 2:
        return "between 1 and 2"
    elif 2 < x <= 3:
        return "between 2 and 3"
    elif 3 < x <= 4:
        return "between 3 and 4"
    elif 4 < x <= 5:
        return "between 4 and 5"
    else:
        return "out of range"


df['rating_obj'] = df['average_rating'].apply(num_to_obj)


print(df['rating_obj'].value_counts())   
```

```python
rating_df = pd.get_dummies(df['rating_obj'])
rating_df.head()
```

```python
df.columns
```

```python
language_df = pd.get_dummies(df['language_code'])
language_df.head()
```

```python
features = pd.concat([rating_df, language_df, df[['average_rating', 'ratings_count', 'title']]], axis=1)

features.set_index('title', inplace=True)
features.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
features_scaled = scaler.fit_transform(features)
```

```python
features_scaled
```

## Model Building 

```python
from sklearn import neighbors 
```

```python
model = neighbors.NearestNeighbors(n_neighbors=5, algorithm = 'ball_tree',
                                  metric = 'euclidean')

model.fit(features_scaled)
dist, idlist = model.kneighbors(features_scaled)
```

```python
df['title'].value_counts()
```

```python
@interact
def BookRecommender(book_name = list(df['title'].value_counts().index)):
    book_list_name = []
    book_id = df[df['title'] == book_name].index
    book_id = book_id[0]
    for newid in idlist[book_id]:
        book_list_name.append(df.iloc[newid].title)
    return book_list_name
        
        
    
```
